# Text Analysis

```python
import collections
import string
import re
from tkinter import *

# Function for clearing the contents of all entry boxes
def clearAll() :
   
    # deleting the content from the entry box
    totalSentences.delete(0, END)
    totalCharacters.delete(0, END)
    totalWords.delete(0, END)
    uniqueWords.delete(0, END)
    specialCharacters.delete(0, END)

# Function for closing window
def exit():
    gui.destroy()

# Function for adding text from file
def AddTxt():
    
    # whole content of text area  is cleared
    textArea.delete(1.0, END)
    
    tf = filedialog.askopenfilename(
        initialdir="C:/Users/Admin/Desktop/", 
        title="Open Text file", 
        filetypes=(("Text Files", "*.txt"),)
        )
    tf = open(tf, 'r', encoding="utf8")
    text = tf.read()
    textArea.insert(END, text)
    

# function for text analysis
def text_analysis():
    clearAll()
    
    # getting input content from text box
    text = textArea.get("1.0", "end")
    
    sen = re.split("\.+", text);
    ts = len(sen) - 1
    totalSentences.insert(END, ts)
    
    tc = 0
    text.replace(" ","")
    for i in range(1, len(text.replace(" ",""))):
        if text[i] != "" :
            tc = tc + 1
        else :
            break
    totalCharacters.insert(END, tc)
    
    counter = collections.Counter(text.split())
    d = counter.most_common()
    tw = sum([i[1] for i in d])
    totalWords.insert(END, tw)
    
    uw = len([i[0] for i in d])
    uniqueWords.insert(END, uw)
    
    sch = string.punctuation
    sc = sum(v for k, v in collections.Counter(text).items() if k in sch)
    specialCharacters.insert(END, sc)


# Driver Code
if __name__ == "__main__" :
    
    # Create a GUI window
    gui = Tk()
    gui.title("Text Analysis")
    gui.geometry("1280x720")
    gui.config(background =  "#65adb2")
    
    #Widgets
    enterText = Label(gui, text = "Enter Your Text", bg = "light green")

    fileBtn = Button(gui, text = "Add text from file", fg = "Black", bg = "#ffcccb", command = AddTxt)
    
    textArea = Text(gui, height = 10, width = 140, font = "consolas")
    
    check = Button(gui, text = "Text Analysis", fg = "Black", bg = "#ffcccb", command = text_analysis)
    
    clear = Button(gui, text = "Clear", fg = "Black", bg = "#ffcccb", command = clearAll)
     
    exit = Button(gui, text = "Exit", fg = "Black", bg = "#ffcccb", command = exit)
    
    total_sentences = Label(gui, text = "Total sentences in text are: ", bg = "light green")
        
    total_characters= Label(gui, text = "Total characters in text are: ", bg = "light green")
    
    total_words = Label(gui, text = "Total words in text are: ", bg = "light green")
    
    unique_words = Label(gui, text = "Unique words in text are: ", bg = "light green")
    
    special_characters = Label(gui, text = "Special characters in text are: ", bg = "light green")
    
    totalSentences = Entry(gui)
    
    totalCharacters = Entry(gui)
    
    totalWords = Entry(gui)
    
    uniqueWords = Entry(gui)
    
    specialCharacters = Entry(gui)
    
 
    # grid method is used for placing the widgets at respective positions in table like structure.

    fileBtn.grid(row = 0, column = 0, padx = 10, pady = 10, columnspan = 3)
     
    textArea.grid(row = 1, column = 0, padx = 10, columnspan= 3, sticky = W,)
     
    check.grid(row = 2, column = 0, padx = 10, pady = 10)
    
    clear.grid(row = 2, column = 1, padx = 10, pady = 10)
     
    exit.grid(row = 2, column = 2, padx = 10, pady = 10)
    
    #grid column 0
    
    total_sentences.grid(row = 3, column = 0, padx = 10, pady = 10)
    
    total_characters.grid(row = 5, column = 0, padx = 10, pady = 10)
     
    total_words.grid(row = 7, column = 0, padx = 10, pady = 10)
     
    unique_words.grid(row = 9, column = 0, padx = 10, pady = 10)
    
    special_characters.grid(row = 11, column = 0, padx = 10, pady = 10)
    
    #grid column 1
    
    totalSentences.grid(row = 3, column = 1)

    totalCharacters.grid(row = 5, column = 1)
                       
    totalWords.grid(row = 7, column = 1)
     
    uniqueWords.grid(row = 9, column = 1)
    
    specialCharacters.grid(row = 11, column = 1)
    
    # start the GUI
    gui.mainloop()
```
